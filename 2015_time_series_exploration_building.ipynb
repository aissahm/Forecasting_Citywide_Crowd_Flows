```
def to_frequency_table(data, frequencytable):
    if frequencytable == None:
        frequencytable = {}
    for key in data:
        keystr = str(key)
        if keystr in frequencytable:
            frequencytable[keystr] += 1
        else:
            frequencytable[keystr] = 1
    return frequencytable

results = None

csvFilePathArray = ["/content/gdrive/My Drive/urban-computing-project/clean-df-jan-start-2015.csv", "/content/gdrive/My Drive/urban-computing-project/clean-df-jan-end-2015.csv", "/content/gdrive/My Drive/urban-computing-project/clean-df-feb-start-2015.csv", "/content/gdrive/My Drive/urban-computing-project/clean-df-feb-end-2015.csv", "/content/gdrive/My Drive/urban-computing-project/clean-df-mar-start-2015.csv", "/content/gdrive/My Drive/urban-computing-project/clean-df-mar-end-2015.csv"]

for csvFilePath in csvFilePathArray:
  df = pd.read_csv(csvFilePath)
  results = to_frequency_table(df['pickup_regionIDs'], results)
  results = to_frequency_table(df['dropoff_regionIDs'], results)

print(results)
```

## **Building the Time Series Matrix**

```
initial_date = datetime.datetime(2015, 1, 1, 0, 0)

def returnIndexForRegionID(regionIDvalue):
  if regionIDvalue in top_16_regions_object:
    return top_16_regions_object[regionIDvalue]
  return -1

def returnHourlyVector(totalhours):
  hourlyVector = [[0, 0]] * totalhours
  return hourlyVector

def returnNullTimeSeriesMatrix():
  number_rows = len(top_16_regions_object.keys())
  initialHourlyVector = returnHourlyVector(24*370*2)
  timeSeriesMatrix = [[[0,0] for j in range(len(initialHourlyVector))] for i in range(number_rows)]
  return timeSeriesMatrix

def convertDatetoHourIndex(taxiTripTimeStamp, yeartaxitrip):
  taxitripdate = datetime.datetime(year=yeartaxitrip, month=taxiTripTimeStamp[0], day=taxiTripTimeStamp[1], hour=taxiTripTimeStamp[2])
  diff = taxitripdate - initial_date
  return int(diff.total_seconds() / 3600.0)

def constructTimeSeriesMatrixFrom2015(dataframe, timeSeriesMatrix, yeartaxitrip):
  
  for index, row in dataframe.iterrows():
    pickupTimestamp = [ row['pickup_month'], row['pickup_day'], row['pickup_hour']]
    pickupHourlyIndex = convertDatetoHourIndex(pickupTimestamp, yeartaxitrip)
    
    pickupIndex = returnIndexForRegionID( row['pickup_regionIDs'] )

    yearadjustment = 0
    if row['pickup_month'] == 12 and row['dropoff_month'] == 1:
      yearadjustment = 1
    else:
      yearadjustment = 0

    dropoffTimestamp = [ row['dropoff_month'], row['dropoff_day'], row['dropoff_hour']]
    dropoffHourlyIndex = convertDatetoHourIndex(dropoffTimestamp, yeartaxitrip+yearadjustment)
    
    dropoffIndex = returnIndexForRegionID( row['dropoff_regionIDs'] )

    if pickupIndex >=0 and dropoffIndex >= 0:
      #pickup update
      currentvalue = timeSeriesMatrix[pickupIndex][pickupHourlyIndex][0]
      timeSeriesMatrix[pickupIndex][pickupHourlyIndex][0] = currentvalue + 1

      #dropoff update
      currentvalue = timeSeriesMatrix[dropoffIndex][dropoffHourlyIndex][1]
      timeSeriesMatrix[dropoffIndex][dropoffHourlyIndex][1] = currentvalue + 1

  return timeSeriesMatrix
```

```
def returnBaseNameOfFileFullPath(resultDatasetPath):
    name = resultDatasetPath.split('/')
    baseName = name[len(name)-1]
    name = baseName.split('.')
    return name[0]
```

```
yearstr_array = ["2015", "2016"]

twoYearsTimeSeriesMatrix = returnNullTimeSeriesMatrix()

base_path = "/content/gdrive/My Drive/urban-computing-project/"

i = 0

for yearstr in yearstr_array:
  for filename in os.listdir("/content/gdrive/My Drive/urban-computing-project/"):
    baseName = returnBaseNameOfFileFullPath(filename)
    if yearstr in baseName and "clean" in baseName:
      print(baseName)
      df_path = base_path + baseName + ".csv"
      df = pd.read_csv(df_path)
      twoYearsTimeSeriesMatrix = constructTimeSeriesMatrixFrom2015(df, twoYearsTimeSeriesMatrix, int(yearstr))
      print(i)
    i = i + 1

print("done")
```

```
from numpy import asarray
from numpy import savetxt
from numpy import loadtxt

def transformTimeSeriesToCSVMatrix(timeSeriesMatrix):
    timeSeriesListMatrix = []
    for timeSeriesArray in timeSeriesMatrix:
        timeSeriesList = []
        for timeSeriesElem in timeSeriesArray:
            timeSeriesList.append(timeSeriesElem[0])
            timeSeriesList.append(timeSeriesElem[1])
        timeSeriesListMatrix.append(timeSeriesList)

    return timeSeriesListMatrix

def saveFileToCSV(timeSeriesMatrix, CSVFilename):
    timeSeriesCSVMatrix = transformTimeSeriesToCSVMatrix(timeSeriesMatrix)
    data = asarray( timeSeriesCSVMatrix )
    # save to csv file
    CSVFullpath = CSVFilename
    savetxt(CSVFullpath, data, delimiter=',')

def returnDataFromCSV(CSVFullpath):
    data = loadtxt(CSVFullpath, delimiter=',')
    return data

csvtwoyearstimeseriespath = "/content/gdrive/My Drive/urban-computing-project/two_years_2015_2016_timeseries.csv"

saveFileToCSV(twoYearsTimeSeriesMatrix, csvtwoyearstimeseriespath)

```

## **Transforming the matrix into a dataframe**

```
def returnOneFlowTimeSeriesArrayFor(regionIDvalue, timeSeriesMatrix, newFlow):
  regionIDindex = returnIndexForRegionID(regionIDvalue)
  if regionIDindex < 0:
    print("Error: region ID not encountered ", regionIDvalue)
    return None
  
  #which flow is requested, True it's new-flow
  # False it's end-flow
  flowIndex = 0
  if newFlow == False:
    flowIndex = 1
  
  flowTimeSeriesArray = []
  n = len(timeSeriesMatrix[regionIDindex])
  i = 0
  while i < n:
    flow_hour_h = timeSeriesMatrix[regionIDindex][i][flowIndex]
    flowTimeSeriesArray.append(flow_hour_h)
    i = i + 1

  return flowTimeSeriesArray
```

```
us_holidays = holidays.UnitedStates()

def returnMonthIndexFor(taxiTripTimeStamp):
  return taxiTripTimeStamp.month

def returnDayIndexFor(taxiTripTimeStamp):
  return taxiTripTimeStamp.day 

def returnWeekdayIndexFor(taxiTripTimeStamp):
  return taxiTripTimeStamp.isoweekday() 

def returnHourIndexFor(taxiTripTimeStamp):
  return taxiTripTimeStamp.hour 

def returnHolidayIndexFor(taxiTripTimeStamp):
  if taxiTripTimeStamp in us_holidays:
    return 1
  return 0

def createDataFrameForTopRegions(timeSeriesMatrix):

  df = pd.DataFrame()
  n = len(twoYearsTimeSeriesMatrix[0])
  date_df = pd.date_range(pd.Timestamp(initial_date),
                    periods=n, freq='h')
  df["timestamp"] = date_df
  df["month"] = df.apply (lambda row: returnMonthIndexFor(row["timestamp"]), axis=1)
  df["day"] = df.apply (lambda row: returnDayIndexFor(row["timestamp"]), axis=1)
  df["hour"] = df.apply (lambda row: returnHourIndexFor(row["timestamp"]), axis=1)
  df["weekday"] = df.apply (lambda row: returnWeekdayIndexFor(row["timestamp"]), axis=1)
  df["holiday"] = df.apply (lambda row: returnHolidayIndexFor(row["timestamp"]), axis=1)

  for regionIDkey in top_16_regions_object:
    df[str(regionIDkey) + "_new"] =  returnOneFlowTimeSeriesArrayFor(regionIDkey, timeSeriesMatrix, True)
    df[str(regionIDkey) + "_end"] = returnOneFlowTimeSeriesArrayFor(regionIDkey, timeSeriesMatrix, False)

  return df
```
